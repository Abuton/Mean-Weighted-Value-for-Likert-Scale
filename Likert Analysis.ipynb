```python
#first let's import the important libraries
import pandas as pd
import numpy as np

path = r"C:\Users\ibrom\Desktop\Data Analysis Spreadsheet\Likert Scale Analysis.xlsx"
df = pd.read_excel(path)
df.head()
```

```python
df.tail()
```

```python
# Coding the categorical data
map = {'Strongly Disagree': 1, 'Disagree':2, 'Agree':3, 'Strongly Agree':4}
df.replace(map, inplace = True)
df
```

```python
#dropping the NaN values
df.dropna(inplace = True)
df.drop("Respondent", axis = 1, inplace = True)
```

```python
df
```

```python
# We need to transpose the dataframe so that the variables becomes the index and number of response is turned to columns
df = df.T(inplace = T)
df
```

```python
# We add four new columns to store the total number of response corresponding to the scale
df['Strongly Disagree'] = (df.iloc[:] == 1).sum(axis =1)
df['Disagree'] = (df.iloc[:] == 2).sum(axis =1)
df['Agree'] = (df.iloc[:] == 3).sum(axis =1)
df['Strongly Agree'] = (df.iloc[:] == 4).sum(axis =1)
df
```

```python
# We only need the newly created columns, so we drop others
df = df.iloc[:, 40:44]
df
```

```python
# We create a new column named Sum of Values to add the scales together
df['SMV'] = df['Strongly Disagree'] + df['Disagree'] + df['Agree'] + df['Strongly Agree']
```

```python
# We define a function to calculate the mean weight
def mwv(a,b,c,d):
    weight = ((a * 1) + (b * 2) + (c * 3) + (d * 4))/(a+b+c+d)
    return weight
# We create a new column to hold the mean weight
df['MWV'] = df.apply( lambda row: mwv(row['Strongly Disagree'], row['Disagree'], row['Agree'], row['Strongly Agree']), axis = 1)
```

```python
# Let's create a new column that ranks the mean weight
df['Rank'] = df['MWV'].rank()
df
```
